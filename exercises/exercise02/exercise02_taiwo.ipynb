```python
# Initialize a counter for request types
request_type_counter = Counter()

# Iterate through each service request and count the request types
for request in service_requests.values():
    request_type = request['Request Type']
    request_type_counter[request_type] += 1

# Print the count of each request type
for request_type, count in request_type_counter.items():
    print(f"Request Type in service_requests {request_type}: {count} times")
```

```python
# Initialize a counter for cities
city_counter = Counter()

# Iterate through each service request and count the cities
for request in service_requests.values():
    city = request['City']
    city_counter[city] += 1

# Print the count of requests made from each city
for city, count in city_counter.items():
    print(f"Requests made from {city}: {count} times")
```
